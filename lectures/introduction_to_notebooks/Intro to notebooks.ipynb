## Markdown

One of the major advantages of notebooks is the ability to interleave code, results and explanatory text. The explanatory text can be edited in the document in the same way as you would edit code. Once executed, it changes to a rendered view. As long as it is in a live notebook, it can be edited at any time by simply double-clicking the cell.

To indicate that a cell is explanatory text, the type of the cell should be changed from the default "code" type, to the "markdown" type. Markdown is the name of the syntax used to write the explanatory text.

### Including equations

In markdown, math can be typeset using LaTeX' math syntax. Inline equations are indicated by surrounding the equation in dollar signs (`$ type equation here $`). For example: $ i\hbar {\frac {\partial }{\partial t}}\Psi (\mathbf {r} ,t)={\hat {H}}\Psi (\mathbf {r} ,t) $. You can also add equation blocks by surrounding the equation(s) with double dollar signs (`$$ type equation(s) here $$`):
$$
i\hbar {\frac {\partial }{\partial t}}\Psi (\mathbf {r} ,t)={\hat {H}}\Psi (\mathbf {r} ,t)
$$

For a full overview of the markdown syntax in notebooks, see http://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/rstversions/Working%20With%20Markdown%20Cells.html

## Plotting

The most popular package in SciPy for creating publication quality figures is matplotlib. The easiest way to use matplotlib is to import the submodule "pyplot". This is heavily inspired by Matlab, and should therefore be easy to get used to if you already have experience with that.

```python
# First line is only needed if you use matplotlib in a Notebook:
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from matplotlib.patches import Polygon

a, b = 2, 9  # integral limits

fig, ax = plt.subplots()
plt.plot(x, y, 'r', linewidth=2)
plt.ylim(ymin=0)

# Make the shaded region
ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

plt.text(0.5 * (a + b), 30, r"$\int_a^b f(x)\mathrm{d}x$",
         horizontalalignment='center', fontsize=20)

plt.figtext(0.9, 0.05, '$x$')
plt.figtext(0.1, 0.9, '$y$')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([])

plt.show()
```

## 3D plots

```python
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=0.3)
cset = ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contourf(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)

ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100)

plt.show()
```
